```python
import numpy as np
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return 1 - np.tanh(x)**2  

inputs = np.array([[0.05, 0.10]])  
expected_output = np.array([[0.01, 0.99]])  

weights = {
    'w1': 0.15, 'w2': 0.20, 'w3': 0.25, 'w4': 0.30,
    'w5': 0.40, 'w6': 0.45, 'w7': 0.50, 'w8': 0.55
}
biases = {'b1': 0.35, 'b2': 0.60}

w = np.array([weights['w1'], weights['w2'], weights['w3'], weights['w4'],
              weights['w5'], weights['w6'], weights['w7'], weights['w8']])
b = np.array([biases['b1'], biases['b2']])
lr = 0.5
epochs = 5000
for epoch in range(epochs):
    h1_input = inputs[0][0] * w[0] + inputs[0][1] * w[2] + b[0]
    h2_input = inputs[0][0] * w[1] + inputs[0][1] * w[3] + b[0]
    
    h1_output = tanh(h1_input)
    h2_output = tanh(h2_input)
    
    o1_input = h1_output * w[4] + h2_output * w[6] + b[1]
    o2_input = h1_output * w[5] + h2_output * w[7] + b[1]
    
    o1_output = tanh(o1_input)
    o2_output = tanh(o2_input)
    
    error_o1 = 0.5 * (expected_output[0][0] - o1_output) ** 2
    error_o2 = 0.5 * (expected_output[0][1] - o2_output) ** 2
    total_error = error_o1 + error_o2
    
    dE_o1 = o1_output - expected_output[0][0]
    dE_o2 = o2_output - expected_output[0][1]
    
    dO1 = dE_o1 * tanh_derivative(o1_input)
    dO2 = dE_o2 * tanh_derivative(o2_input)
    
    dW5 = dO1 * h1_output
    dW6 = dO1 * h2_output
    dW7 = dO2 * h1_output
    dW8 = dO2 * h2_output
    
    db2 = dO1 + dO2
    
    dH1 = dO1 * w[4] + dO2 * w[6]
    dH2 = dO1 * w[5] + dO2 * w[7]
    
    dH1_input = dH1 * tanh_derivative(h1_input)
    dH2_input = dH2 * tanh_derivative(h2_input)
    
    dW1 = dH1_input * inputs[0][0]
    dW2 = dH1_input * inputs[0][1]
    dW3 = dH2_input * inputs[0][0]
    dW4 = dH2_input * inputs[0][1]
    
    db1 = dH1_input + dH2_input
    
    w[0] -= lr * dW1
    w[1] -= lr * dW2
    w[2] -= lr * dW3
    w[3] -= lr * dW4
    w[4] -= lr * dW5
    w[5] -= lr * dW6
    w[6] -= lr * dW7
    w[7] -= lr * dW8
    
    b[0] -= lr * db1
    b[1] -= lr * db2

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}: Total Error = {total_error:.6f}")

print("Final Outputs:", o1_output, o2_output)
```
